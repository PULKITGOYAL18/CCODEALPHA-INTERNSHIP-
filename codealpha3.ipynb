```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train[0].shape
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
x_train=x_train/255#normalizing the pixel values
```

```python
x_test=x_test/255
```

```python
x_train[0]
```

```python
x_train2 = x_train.reshape(len(x_train), 28*28)
x_test2 = x_test.reshape(len(x_test), 28*28)
```

```python
x_train2.shape
```

```python
x_train2[0]
```

```python
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(784,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')  # 10 classes for digits 0–9
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train2, y_train, epochs=10, validation_data=(x_test2, y_test))
```

```python
model.evaluate(x_test2,y_test)
```

```python
y_predicted = model.predict(x_test2)
y_predicted[0]
```

```python
plt.matshow(x_test[0])
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train2, y_train, epochs=5)
```

```python
model.evaluate(x_test2,y_test)
```

```python
y_predicted = model.predict(x_test2)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```
